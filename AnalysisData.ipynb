```python
%matplotlib inline 
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import tmdbsimple as tmdb
from imdb import IMDb
import numpy as np
import pandas as pd
import time
import cPickle as pickle
import io
```

```python
MovieTable1 = pd.read_pickle('MovieData100_10_15.pkl')
```

```python
MDdf = pd.DataFrame(MovieTable1).transpose()
MDdf.head()
```

```python
MD_director_gross = MDdf[['director', 'gross']]
```

```python
director_list = list()
for dl in MD_director_gross.director.get_values():
    if dl == None:
        continue
    for d in dl:
        if d not in director_list:
            director_list.append(d)
```

```python
director_list
```

```python
dd = dict()
for d in director_list:
    tl = list()
    for gg in MD_director_gross.director:
        if gg == None:
            tl.append(False)
        else:
            tl.append(d in gg)
    dd[d] = MD_director_gross[tl].gross
```
